```python
import pandas as pd
```

### Importanto o dataset do Github para dentro da base!

```python
#para url do github tem que adicionar o ?raw=true
url_categories = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Categories.csv?raw=true'
url_customers = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Customers.csv?raw=true'
url_employeestatus = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/EmployeeStatus.csv?raw=true'
url_employess = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Employees.csv?raw=true'
url_ordersdetails_01 = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/OrderDetails1.csv?raw=true'
url_orderdetaisls_02 = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/OrderDetails2.csv?raw=true'
url_orders_01 = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Orders1.csv?raw=true'
url_orders_02 = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Orders2.csv?raw=true'
url_products = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Products.csv?raw=true'
url_regions = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Regions.csv?raw=true'
url_shippers = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Shippers.csv?raw=true'
url_suppliers = 'https://github.com/fabiobatistapinheiro/dunder_mifflin/blob/Dataset/Suppliers.csv?raw=true'
categories = pd.read_csv(url_categories,index_col=0, sep=';')
customers = pd.read_csv(url_customers,index_col=0, sep=';', encoding='latin')
employeestatus = pd.read_csv(url_employeestatus,index_col=0, sep=';')
ordersdetails_01 = pd.read_csv(url_ordersdetails_01,index_col=0, sep=';')
orderdetaisls_02 = pd.read_csv(url_orderdetaisls_02,index_col=0, sep=';')
orders_01 = pd.read_csv(url_orders_01,index_col=0, sep=';', encoding='latin')
orders_02 = pd.read_csv(url_orders_02,index_col=0, sep=';')
products = pd.read_csv(url_products,index_col=0, sep=';')
regions = pd.read_csv(url_regions,index_col=0, sep=';')
shippers = pd.read_csv(url_shippers,index_col=0, sep=';')
suppliers = pd.read_csv(url_suppliers,index_col=0, sep=';')
```

#### Concatenamos os pedidos e seus detalhes em um único dataset

```python
orders = pd.concat([orders_01, orders_02], axis=0)
ordersdetails = pd.concat([ordersdetails_01, orderdetaisls_02], axis=0)
```
